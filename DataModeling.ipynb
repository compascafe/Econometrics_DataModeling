```python
import pandas as pd

import matplotlib.pyplot as plt

# Load your VIX dataset (replace 'vix_data.csv' with your actual file path)
df = pd.read_csv('datasets/VIX.csv')

# Make sure the 'Date' column is in datetime format and set it as the index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Sort the DataFrame by date (if it's not already sorted)
df = df.sort_index()

# Calculate daily returns
df['Daily_Return'] = df['Close'].pct_change()




# Plotting the daily returns
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Daily_Return'], label='Daily Returns', color='blue')
plt.title('VIX Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily Returns')
plt.legend()
plt.grid(True)
plt.show()
```
