```python
import pandas as pd

import matplotlib.pyplot as plt

# Load your VIX dataset (replace 'vix_data.csv' with your actual file path)
df = pd.read_csv('datasets/VIX.csv')

# Make sure the 'Date' column is in datetime format and set it as the index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Sort the DataFrame by date (if it's not already sorted)
df = df.sort_index()

# Calculate daily returns
df['Daily_Return'] = df['Close'].pct_change()




# Plotting the daily returns
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Daily_Return'], label='Daily Returns', color='blue')
plt.title('VIX Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily Returns')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

stock_symbols = ['VIX', 'SPY', 'DIA', 'ONEQ']

plt.figure(figsize=(12, 8))

# Plot each stock on the same plot
for stock_symbol in stock_symbols:
    # Load stock data (replace 'stock_symbol.csv' with the actual file path)
    stock_data = pd.read_csv(f'datasets/{stock_symbol}.csv')

    # Convert 'Date' column to datetime format and set it as the index
    stock_data['Date'] = pd.to_datetime(stock_data['Date'])
    stock_data.set_index('Date', inplace=True)

    # Sort the DataFrame by date (if it's not already sorted)
    stock_data = stock_data.sort_index()

    # Plotting the stock prices
    plt.plot(stock_data.index, stock_data['Close'], label=f'{stock_symbol} Close Price')

# Set labels and title
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
